```python
#Intro to Matplotlib
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
%matplotlib inline
```

```python
plt.plot([1,2,3,4], [11,22,33,44]);
plt.show();
```

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
#plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
#make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show();
```
