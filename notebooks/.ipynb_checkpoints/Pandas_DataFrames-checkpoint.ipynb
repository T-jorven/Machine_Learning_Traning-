```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/heart-disease.csv")
```

```python
over_50 = df[df["age"] > 50]
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6));

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="coolwarm");

# Customize the plot

ax.set(title= "Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend 
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line for the average values
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# ubplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10));

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="coolwarm");

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline
ax0.axhline(over_50["chol"].mean(), linestyle="--");




# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="coolwarm");

# Customize ax1
ax1.set(title="Heart Disease and Heart Rate Levels",
       xlabel="Age",
       ylabel="Heart Rate");

# Add a legend to ax0
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a meanline
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

# Add a title to the figute
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
