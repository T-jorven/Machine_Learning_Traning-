```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
%matplotlib inline
```

```python
# Create a simple plot using plt.plot()

# Example data (time vs. equity)
time = [1, 2, 3, 4, 5]
equity = [1000, 1050, 1030, 1080, 1100]

# Create plot
plt.plot(time, equity, label="Equity Curve", color="blue", linewidth=2)

# Add labels and title
plt.xlabel("Time (Days)")
plt.ylabel("Equity ($)")
plt.title("EA Equity Curve")
plt.legend()

# Show plot
plt.show()
```

```python
# Plot a single Python list

# Single list of values
values = [3, 7, 4, 5, 9, 2]

# Plot
plt.plot(values, marker="o", linestyle="-", color="b")

# Labels and title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Plot of a Single List")

plt.show()
```

```python
# Create two lists, one called X, one called y, each with 5 numbers in them



# Two lists
X = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot
plt.plot(X, y, marker="o", linestyle="-", color="r")

# Labels and title
plt.xlabel("X values")
plt.ylabel("y values")
plt.title("Plot of X vs y")

plt.show()
```

```python
# Create a plot using plt.subplots() and then add X & y on the axes
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot on the axes
ax.plot(x, y, marker="o", linestyle="-", color="g")

# Labels and title
ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Plot using OO method")

# Show the plot
plt.show()
```

```python
#Now let's try a small matplotlib workflow.

# Prepare data (create two lists of 5 numbers, X & y)
X = [2,1,6,4,7]
y = [9,3,4,1,5]

# Setup figure and axes using plt.subplots()

fig, ax = plt.subplots()

# Add data (X, y) to axes

ax.plot(X, y)

# Customize plot by adding a title, xlabel and ylabel

ax.plot(X, y, marker="o", linestyle="-", color="g")


# Save the plot to file using fig.savefig()

fig.savefig("my_plot.png");

plt.show();
```

```python
# Create an array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X

X = np.linspace(0, 100, 100)
Y = X ** 2

# Create a plot using plt.subplots() and plot X versus X^2 (X squared)

fig, ax = plt.subplots()
ax.plot(X, Y, color="blue", label="X squared")

ax.set_xlabel("X")
ax.set_ylabel("X²")
ax.set_title("Plot of X vs X²")
ax.legend()

fig.savefig("x_squared_plot.png", dpi=300)

plt.show();
```

```python
#We'll start with scatter plots.

# Create a scatter plot of X versus the exponential of X (np.exp(X))

X = np.linspace(0, 100, 100)

# Create a scatter plot of X versus np.sin(X)

# --- Scatter: X vs exp(X) ---
fig1, ax1 = plt.subplots()
ax1.scatter(X, np.exp(X))
ax1.set_xlabel("X")
ax1.set_ylabel("exp(X)")
ax1.set_title("Scatter: X vs exp(X)")

# --- Scatter: X vs sin(X) ---
fig2, ax2 = plt.subplots()
ax2.scatter(X, np.sin(X))
ax2.set_xlabel("X")
ax2.set_ylabel("sin(X)")
ax2.set_title("Scatter: X vs sin(X)")

plt.show()
```
